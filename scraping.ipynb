```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
#Example using "mira"
url = "https://en.m.wiktionary.org/wiki/mira#Etymology"
source_code = requests.get(url)
plain_text = source_code.text
```

```python
soup = BeautifulSoup(plain_text, 'html.parser')
columns = soup.find('div', {'class': 'content'}).find('div', {'class': 'mw-content-ltr'}).find('div', {'class': 'mw-parser-output'}).findAll('section')
for column in columns:
    lang = column.select("p a")
    lang = list(lang)
    for x in lang:
        x = list(x)
        print(x)
   
```

```python
#alternative way using the API
import requests

S = requests.Session()

URL = "https://en.m.wiktionary.org/w/api.php"

PARAMS = {
    "action": "parse",
    "page": "laude",
    "format": "json"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

print(DATA["parse"]["text"]["*"])
```
